```python
import numpy as np
import pandas as pd
import scipy as sp
import time
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from DataServer import DataServer
from GenerateScenarios import GenerateScenarios
from PricingDerivatives import PricingDerivatives
from HedgingModel import HedgingModel
```

```python
dataServer = DataServer()
pricingDerivatives = PricingDerivatives()
generateScenarios = GenerateScenarios(dataServer)
hedgingModel = HedgingModel(dataServer, pricingDerivatives)
```

## Computational time comparison

```python
num_scenarios = 10000
alpha = 0.05

f_list = []
x_list = []
t_list = []

# generate scenarios
p_T_scenarios = generateScenarios.get_price_scenarios(num_scenarios)
V_T_scenarios = generateScenarios.get_volume_scenarios(num_scenarios)

# matricial form model
start_time = time.time()
f, x = hedgingModel.opt_hedging_model_matrices(p_T_scenarios=p_T_scenarios, V_T_scenarios=V_T_scenarios, alpha=alpha, show_results=False)
end_time = time.time()
t = end_time-start_time
f_list.append(f)
x_list.append(x)
t_list.append(t)

# for loops model
start_time = time.time()
f, x = hedgingModel.opt_hedging_model_for_loops(p_T_scenarios=p_T_scenarios, V_T_scenarios=V_T_scenarios, alpha=alpha, show_results=False)
end_time = time.time()
t = end_time-start_time
f_list.append(f)
x_list.append(x)
t_list.append(t)

# comparison dataframe
df_comparison = pd.DataFrame(x_list, index=['matricial form', 'for loops'])
df_comparison['f'] = f_list
df_comparison['t'] = t_list
```

## Stability analysis

We use the following block, varying the parameter num_scenarios, to generate different files named 'df_stability_{num_scenarios}_100.csv'.

Each file contains, in row ùëñ, the optimal CVaR value and the optimal first-stage decision for the ùëñ-th replication.

```python
num_rep = 100
num_scenarios = 1000
alpha = 0.05

f_list = []
x_list = []

for i in range(num_rep):

    print(f'\r current replication: {i+1}/{num_rep}', end='')

    p_T_scenarios = generateScenarios.get_price_scenarios(num_scenarios)
    V_T_scenarios = generateScenarios.get_volume_scenarios(num_scenarios)

    f, x = hedgingModel.opt_hedging_model_matrices(p_T_scenarios=p_T_scenarios, V_T_scenarios=V_T_scenarios, alpha=alpha, show_results=False)

    f_list.append(f)
    x_list.append(x)

df_stability = pd.DataFrame(x_list)
df_stability['f'] = f_list
```

## Risk adversion analysis

We use the following block to generate the file named 'df_adversion_20000.csv'.

The file contains, in each row, the optimal CVaR value and the optimal first-stage decision for a different value of the parameter alpha.

```python
alpha_vec = [0.01, 0.05, 0.10, 0.15, 0.20, 0.25]
num_scenarios = 20000

p_T_scenarios = generateScenarios.get_price_scenarios(num_scenarios)
V_T_scenarios = generateScenarios.get_volume_scenarios(num_scenarios)

f_list = []
x_list = []

for alpha in alpha_vec:

    print(f'\r current alpha: {alpha:.2f} \n', end='')

    f, x = hedgingModel.opt_hedging_model_matrices(p_T_scenarios=p_T_scenarios, V_T_scenarios=V_T_scenarios, alpha=alpha, show_results=False)
    f_list.append(f)
    x_list.append(x)

df_adversion = pd.DataFrame(x_list, index=alpha_vec)
df_adversion['f'] = f_list
```

## Uncertainty grade analysis: price volatility


We use the following block to generate the file named 'df_uncertainty_sigma_20000.csv'.

The file contains, in each row, the optimal CVaR value and the optimal first-stage decision for a different value of the price volatility.

```python
num_scenarios = 20000
alpha = 0.05
sigma_vec = [0.10, 0.15, 0.20, 0.25, 0.30]

f_list = []
x_list = []

for sigma in sigma_vec:
    
    print(f'\r current sigma: {sigma:.2f} \n', end='')

    dataServer = DataServer(sigma=sigma)
    pricingDerivatives = PricingDerivatives()
    hedgingModel = HedgingModel(dataServer, pricingDerivatives)
    generateScenarios = GenerateScenarios(dataServer)

    p_T_scenarios = generateScenarios.get_price_scenarios(num_scenarios)
    V_T_scenarios = generateScenarios.get_volume_scenarios(num_scenarios)

    f, x = hedgingModel.opt_hedging_model_matrices(p_T_scenarios=p_T_scenarios, V_T_scenarios=V_T_scenarios, alpha=alpha, show_results=False)

    f_list.append(f)
    x_list.append(x)

df_uncertainty_sigma = pd.DataFrame(x_list, index=sigma_vec)
df_uncertainty_sigma['f'] = f_list
```

```python
df_uncertainty_sigma
```

## Uncertainty grade analysis: volume

We use the following block to generate the file named 'df_uncertainty_vol_20000.csv'.

The file contains, in each row, the optimal CVaR value and the optimal first-stage decision for a different volume distribution, in particular with different standard deviations.

```python
num_scenarios = 20000
alpha = 0.05
std_vec = []
V_dist_vec = [
    sp.stats.triang(c=0.75, loc=17500*1000, scale=5000*1000),
    sp.stats.triang(c=0.75, loc=15000*1000, scale=10000*1000),
    sp.stats.triang(c=0.75, loc=10000*1000, scale=20000*1000),
    sp.stats.triang(c=0.75, loc=5000*1000, scale=30000*1000),
    sp.stats.triang(c=0.75, loc=2500*1000, scale=37500*1000)
]

f_list = []
x_list = []

for i in range(len(V_dist_vec)):
    
    V_i = V_dist_vec[i]
    dataServer = DataServer(V_dist = V_i)
    pricingDerivatives = PricingDerivatives()
    hedgingModel = HedgingModel(dataServer, pricingDerivatives)
    generateScenarios = GenerateScenarios(dataServer)

    p_T_scenarios = generateScenarios.get_price_scenarios(num_scenarios)
    V_T_scenarios = generateScenarios.get_volume_scenarios(num_scenarios)

    std_vec.append(np.std(V_T_scenarios))
    print(f'\r current std: {np.std(V_T_scenarios):.2f} \n', end='')

    f, x = hedgingModel.opt_hedging_model_matrices(p_T_scenarios=p_T_scenarios, V_T_scenarios=V_T_scenarios, alpha=alpha, show_results=False)

    f_list.append(f)
    x_list.append(x)

df_uncertainty_vol = pd.DataFrame(x_list, index=std_vec)
df_uncertainty_vol['f'] = f_list
```

## Out of sample stability

```python
dataServer = DataServer()
pricingDerivatives = PricingDerivatives()
generateScenarios = GenerateScenarios(dataServer)
hedgingModel = HedgingModel(dataServer, pricingDerivatives)
```

```python
num_scenarios = 20000
p_T_scenarios_in = generateScenarios.get_price_scenarios(num_scenarios)
V_T_scenarios_in = generateScenarios.get_volume_scenarios(num_scenarios)
```

```python
alpha = 0.05
cvar, opt_sol = hedgingModel.opt_hedging_model_matrices(p_T_scenarios=p_T_scenarios_in, V_T_scenarios=V_T_scenarios_in, alpha=alpha, show_results=True)
```

```python
num_scenarios_out = 2000
p_T_scenarios_out = generateScenarios.get_price_scenarios(num_scenarios_out)
V_T_scenarios_out = generateScenarios.get_volume_scenarios(num_scenarios_out)
```

```python
loss_usd_in, exercised_call_eur_in, delta_eur_in = hedgingModel.evaluate_solution(opt_sol, p_T_scenarios_in, V_T_scenarios_in)
loss_usd_out, exercised_call_eur_out, delta_eur_out = hedgingModel.evaluate_solution(opt_sol, p_T_scenarios_out, V_T_scenarios_out)
```

```python
plt.figure()
plt.hist(loss_usd_in, bins=50, alpha=0.6, label='in_sample', density=True)
plt.hist(loss_usd_out, bins=50, alpha=0.6, label='out_sample', density=True)
plt.legend()
plt.savefig('out_of_sample_eval.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
